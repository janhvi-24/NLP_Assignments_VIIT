```python
text = "NLTK is amazing! It helps in tokenization, stemming, and lemmatization. Let's test it #NLP"
```

```python
import nltk
from nltk.tokenize import word_tokenize, TweetTokenizer, MWETokenizer
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer
import string

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
whitespace_tokens = text.split()
print("Whitespace Tokenization:")
print(whitespace_tokens)
```

```python
no_punct_text = text.translate(str.maketrans('', '', string.punctuation))
punct_tokens = no_punct_text.split()

print("\nPunctuation-Based Tokenization:")
print(punct_tokens)
```

```python
treebank_tokenizer = TreebankWordTokenizer()
treebank_tokens = treebank_tokenizer.tokenize(text)

print("\nTreebank Tokenization:")
print(treebank_tokens)
```

```python
tweet_tokenizer = TweetTokenizer()
tweet_tokens = tweet_tokenizer.tokenize(text)

print("\nTweet Tokenization:")
print(tweet_tokens)
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
mwe_tokenizer = MWETokenizer([('machine', 'learning'), ('natural', 'language', 'processing')], separator='_')
mwe_tokens = mwe_tokenizer.tokenize(word_tokenize(text))

print("\nMWE Tokenization:")
print(mwe_tokens)
```

```python
porter = PorterStemmer()
porter_stems = [porter.stem(word) for word in word_tokenize(text)]

print("\nPorter Stemmer Output:")
print(porter_stems)
```

```python
snowball = SnowballStemmer("english")
snowball_stems = [snowball.stem(word) for word in word_tokenize(text)]

print("\nSnowball Stemmer Output:")
print(snowball_stems)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in word_tokenize(text)]

print("\nLemmatization Output:")
print(lemmatized_words)
```
